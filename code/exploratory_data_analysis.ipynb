# Data Analysis
Our data consists of Japanese restaurants and our goal is to predict the future amount of visitors each restaurant will receive.  With this, restaurants can better plan inventory and staff accordingly to the forecasted amount of visitors.  Also, visitors could use this forecast to plan ahead before going out to eat.  They can see if a restaurant expects to be busy and if they should make reservations ahead of time to ensure they get a seat. All 7 files are relational to each other by the columns `air_store_id` and `hpg_store_id`.  They consist of user information collected from two Japanese websites, "Hot Pepper Gourmet (hpg): similar to Yelp” (search and reserve) and “AirREGI / Restaurant Board (air): similar to Square” (reservation control and cash register). 
In this notebook we will do exploratory data analysis to better understand trends and relations in our data by the following steps
1. Individual Feature Relations
2. Feature Relations

### Air Visits
First, let's look at the dataframe `air_visit_data`.  This dataframe consists of the restaurants id, date, and how many visitors the resaurant received on each date. There's no missing data and we have 829 unique restaurants in Air's system

```python
total_visitors = air_visit_data.groupby(['visit_date']).sum()
total_visitors.reset_index(inplace=True)
total_visitors['month'] = total_visitors['visit_date'].dt.month
total_visitors['day'] = total_visitors['visit_date'].dt.weekday
```

```python
def num_to_month(df, column):
    months = []
    for row in df[column]:
        if row == 1:
            months.append('Jan')
        elif row == 2:
            months.append('Feb')
        elif row == 3:
            months.append('Mar')
        elif row == 4:
            months.append('Apr')
        elif row == 5:
            months.append('May')
        elif row == 6:
            months.append('Jun')
        elif row == 7:
            months.append('Jul') 
        elif row == 8:
            months.append('Aug')
        elif row == 9:
            months.append('Sep')
        elif row == 10:
            months.append('Oct')
        elif row == 11:
            months.append('Nov')
        elif row == 12:
            months.append('Dec')
    df[column] = pd.Series(months)
```

```python
def num_to_weekday(df, column):
    days = []
    for row in df[column]:
        if row == 0:
            days.append('Mon')
        elif row == 1:
            days.append('Tue')
        elif row == 2:
            days.append('Wed')
        elif row == 3:
            days.append('Thu')
        elif row == 4:
            days.append('Fri')
        elif row == 5:
            days.append('Sat')
        elif row == 6:
            days.append('Sun')
    df[column] = pd.Series(days)
```

```python
num_to_month(total_visitors,'month')
num_to_weekday(total_visitors,'day')
```

```python
fig, ax = plt.subplots(figsize=(12,5))

ax.plot(total_visitors['visit_date'], total_visitors['visitors'], c='darkgreen')
ax.grid(b=True, which='major', color='#DDDDDD', linestyle='-')
ax.set_xlabel('Date')
ax.set_ylabel('Visitors')
plt.title('Total Visitors');
```

```python
median_visitors_month = total_visitors.groupby(['month']).median()
median_visitors_month.reset_index(inplace=True)

median_visitors_day = total_visitors.groupby(['day']).median()
median_visitors_day.reset_index(inplace=True)
```

```python
plt.subplots_adjust(left=2,right=4)
plt.subplot(1,2,1)
sns.set()
sns.barplot(median_visitors_month['month'], median_visitors_month['visitors'], order=['Jan','Feb','Mar','Apr',
                                            'May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'], palette='BuGn_d')
plt.title('Median Visitors per Month')

plt.subplot(1,2,2)
sns.set()
sns.barplot(median_visitors_day['day'], median_visitors_day['visitors'], order=['Sun','Mon',
                                            'Tue','Wed','Thu','Fri','Sat'], palette='BuGn_d')
plt.title('Median Visitors per Day');
```

The time series plot shows a periodic pattern most likely stemming from the weekly cycle.  We see a sharp drop on new years maybe because of holidays.  However, there seems to be a big jump in number of visitors after July 2016.  This could be from an increase of new restaurants in Air's systems.  Also, this jump in visitors might be skewing our graphs of the median amount of visitors there are each month.  The months from January - June all have a lot lower median visitors value than July - December.  December seems to be the most popular month to eat out.  Our weekly median visitors value reinforce our domain knowledge.  Friday and Saturday are the most popular for reservations while Monday is the least.

### Holiday Impact
Lets see if there exists any relations between visitors and reservations on holidays. Let's merge the two dataframes `air_visit_data` and `date_info` as well as `air_reserve` and `date_info`

```python
date_info['visit_date'] = date_info['calendar_date']
holiday_visitors = pd.merge(air_visit_data, date_info, on='visit_date', how='inner')
holiday_reservations = pd.merge(air_reserve, date_info, on='visit_date', how='inner')
```

```python
plt.subplots_adjust(left=2,right=3.1)

plt.subplot(1,2,1)
sns.set()
sns.barplot(holiday_visitors.groupby(['holiday_flg']).median().index.tolist(), 
            holiday_visitors.groupby(['holiday_flg']).median()['visitors'], palette='seismic')
plt.xlabel('Holiday')
plt.xticks(ticks=[0,1],labels=['False', 'True'])
plt.title('Holiday Visitors');

plt.subplot(1,2,2)
sns.set()
sns.barplot(holiday_reservations.groupby(['holiday_flg']).median().index.tolist(), 
            holiday_reservations.groupby(['holiday_flg']).median()['reserve_visitors'],
                                                                            palette='seismic')
plt.ylabel('Reservation visitors')
plt.xlabel('Holiday')
plt.xticks(ticks=[0,1],labels=['False', 'True'])
plt.title('Holiday Reservations');
```

So holidays overall do not seem to make much of a difference in the median visitors and has no impact on the median reservation visitors. Let's see the amount of visitors and reservations for each holiday to see if some holidays have a bigger impact than others

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(holiday_visitors.loc[(holiday_visitors['holiday_flg']==1)].groupby(['visit_date']).median().index.tolist(),
                 holiday_visitors.loc[(holiday_visitors['holiday_flg']==1)].groupby(['visit_date']).median()['visitors'],
                                                                                                        palette='viridis')
plt.xticks(rotation=70)
plt.yticks([0,5,10,15,17,20,25])
plt.ylabel('Visitors', fontsize=15)
plt.title('Holidays', fontsize=20)
plt.plot(ax.get_xticks(), [17]* 31, c='red', label='Median Visitors Non-Holiday')
plt.legend();
```

Here I've plotted the median amount of visitors for each holiday with the red line which represents the median visitors among non-holidays. We can see that December 31st and Jan 1st have a negative impact on visitors wheras Jan 2nd-3rd, Feb 11th, Mar 20th, April, 29th, May 3rd-4th, Dec 23rd, and Dec 29th-30th all have positive impacts.  The other holidays do not make a significant difference
